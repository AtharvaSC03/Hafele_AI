```python
d = df[['product_name', 'sku', 'category', 'description', 'image_url', 'Room Type', 'Themes']]
d.columns = ['product_name', 'sku', 'category', 'description', 'image_url', 'room_type', 'themes']
d.head()
```

```python
for col in ['product_name','category', 'description','room_type', 'themes']:
  d[col] = d[col].fillna("").str.lower().str.strip()
```

```python
# Create a combined text field for ML
d["combined_text"] = (
    d["product_name"] + " " +
    d["category"] + " " +
    d["themes"] + " " +
    d["room_type"] + " " +
    d["description"]
)
```

```python
d.head()
```

TF-IDF (Term Frequency–Inverse Document Frequency) turns each product description into a vector of numbers representing important words.

stop_words="english" removes useless words like “the”, “and”, etc.

max_features=5000 means we only keep the top 5,000 most important words.

tfidf_matrix is the actual numerical representation of all products.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF vectorizer (ignore common words, limit to 5000 features)
vectorizer = TfidfVectorizer(stop_words = 'english', max_features = 5000)
# Fit the model on our combined_text
x = vectorizer.fit_transform(d['combined_text'])  # tf_idf matrix

print('TF_IDF shape',x.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
def search_products(query, top_n = 5):
  #convert query to TF_IDF vector
  query_vec = vectorizer.transform([query])

  # calculate the cosine similarity btw the query and all products
  similarities = cosine_similarity(query_vec,x).flatten()

  # get top N
  top_i = similarities.argsort()[::-1][:top_n]

  # return top products
  return df.iloc[top_i][["product_name", "sku", "category", "description"]]

# Example serach
result = search_products('Modern kitchen lighting', top_n = 5)
print(result)
```
